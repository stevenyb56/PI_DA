```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

En este notebook se realizaran ualgunas transformaciones para poder aplicar los kpis correspondientes.

```python
df = pd.read_csv("../data/limpios/Acceso_x_Provincia_100Hogares.csv")
df
```

```python
df['Provincia'].unique()
```

```python
# Se tomaran los datos del año 2022 
df = df[df['Año'] >= 2022]
```

```python
# Supongamos que tu DataFrame se llama df
# Filtrar el DataFrame para obtener solo los trimestres 3 y 4
trim_3 = df[df['Trimestre'] == 3]
trim_4 = df[df['Trimestre'] == 4]

# Fusionar los DataFrames en función de las columnas 'Año', 'Provincia'
df_resultado = pd.merge(trim_3, trim_4, on=['Año', 'Provincia'], suffixes=(' 3', ' 4'))


# Se cambia el tipo de dato y se redondea
df_resultado['Accesos por cada 100 hogares 3'] = df_resultado['Accesos por cada 100 hogares 3'].round().astype(int)
df_resultado['Accesos por cada 100 hogares 4'] = df_resultado['Accesos por cada 100 hogares 4'].round().astype(int)

```

```python
df_resultado
```

```python
# Se guardas el Df en un formato .csv
carpeta = '../data/limpios/'
nombre ='pki1.csv'
x = carpeta + nombre

df_resultado.to_csv(x, index=False)
```

```python
df2 = pd.read_csv("../data/limpios/Conexiones_x_Provincia.csv")
df2
```

```python
# Especifica el año y trimestre que te interesa
año_interes = 2022
trimestre_interes = 3

# Filtra el DataFrame según el año y trimestre
df_1= df2[(df2['Año'] == año_interes) & (df2['Trimestre'] == trimestre_interes)]

# Selecciona solo las columnas relevantes (Provincia y Fibra óptica)
df_resultado = df_1[['Provincia', 'Fibra óptica']]
df_resultado
```

```python
# Especifica el año y trimestre que te interesa
año_interes = 2022
trimestre_interes = 4

# Filtra el DataFrame según el año y trimestre
df_1= df2[(df2['Año'] == año_interes) & (df2['Trimestre'] == trimestre_interes)]

# Selecciona solo las columnas relevantes (Provincia y Fibra óptica)
df_resultado = df_1[['Provincia', 'Fibra óptica']]
df_resultado
```

```python
# Filtrar solo las filas correspondientes al año 2022, trimestres 1 y 2, y las columnas de interés
df_filt = df2[(df2['Año'] == 2022) & (df2['Trimestre'].isin([1, 2]))]

# Agrupar por 'Año', 'Provincia' y sumar los valores para las columnas de interés
df_agrupado = df_filt.groupby(['Año', 'Provincia']).agg({
    'Cablemodem': 'sum',
    'Fibra óptica': 'sum'
}).reset_index()

# Renombrar las columnas si es necesario
df_agrupado = df_agrupado.rename(columns={'Cablemodem': 'Cablemodem_Semestre1', 'Fibra óptica': 'Fibra_optica_Semestre1'})
df_agrupado
# Fusionar con el DataFrame original
#df_resultado = pd.merge(df, df_agrupado, on=['Año', 'Provincia'], how='left')
```

```python
# Filtrar solo las filas correspondientes al año 2022, trimestres 1 y 2, y las columnas de interés
df_filt = df2[(df2['Año'] == 2022) & (df2['Trimestre'].isin([3, 4]))]

# Agrupar por 'Año', 'Provincia' y sumar los valores para las columnas de interés
df_agrupa = df_filt.groupby(['Año', 'Provincia']).agg({
    'Cablemodem': 'sum',
    'Fibra óptica': 'sum'
}).reset_index()

# Renombrar las columnas si es necesario
df_agrupado2 = df_agrupa.rename(columns={'Cablemodem': 'Cablemodem_Semestre2', 'Fibra óptica': 'Fibra_optica_Semestre2'})
df_agrupado2
# Fusionar con el DataFrame original
#df_resultado = pd.merge(df, df_agrupado, on=['Año', 'Provincia'], how='left')
```

```python
# Se unen los dos DF
df_final = df_agrupado.merge(df_agrupado2, on=['Año', 'Provincia'], how='inner')
orden_columnas = ['Año', 'Provincia', 'Cablemodem_Semestre1', 'Cablemodem_Semestre2', 'Fibra_optica_Semestre1', 'Fibra_optica_Semestre2']
df_final = df_final[orden_columnas]
df_final
```

```python
# Se guardas el Df en un formato .csv
carpeta = '../data/limpios/'
nombre ='pki2.csv'
x = carpeta + nombre

df_final.to_csv(x, index=False)
```

```python
# Se verifican los porcentajes
df_final['Porcentaje_cable'] = (((df_final['Cablemodem_Semestre2'] - df_final['Cablemodem_Semestre1']) / df_final['Cablemodem_Semestre1']) * 100).round(2)
df_final['Porcentaje_fibra'] = (((df_final['Fibra_optica_Semestre2'] - df_final['Fibra_optica_Semestre1']) / df_final['Fibra_optica_Semestre1']) * 100).round(2)
```
