Tomando como ejemplo la localidad 'Barrio Empleados de Comercio' en la provincia de Mendoza, departamento de San Rafael, una revisión en Google Maps revela que se trata de un barrio situado a cierta distancia de San Rafael, pero no lo suficientemente remoto como para carecer de cualquier forma de tecnología, como telefonía fija o cobertura 3G/4G. En consecuencia, se procederá a eliminar las filas que no dispongan de ninguna tecnología con el objetivo de mejorar la integridad de los datos de manera precisa.


```python
# Depuracion de datos faltantes
tecnologias = ['ADSL', 'Cablemódem', 'Dial Up', 'Fibra Óptica', '4G', '3G', 'Telefonía Fija', 'Wireless', 'Satelital']

# Se crea una máscara booleana para identificar las filas con al menos una tecnología ('Si')
mascara_tecnologias = (Loc_con_int[tecnologias] == 'Si')

# Se crea otra una máscara para identificar las filas que tienen al menos una tecnología
mascara_filas_con_tecnologia = mascara_tecnologias.any(axis=1)

# Se filtra el DataFrame utilizando la máscara anterior
DF_Loc_con_int = Loc_con_int[mascara_filas_con_tecnologia]
```

```python
# Se convierte las tecnologia a forma numerica
tecnologias = ['ADSL', 'Cablemódem', 'Dial Up', 'Fibra Óptica', '4G', '3G', 'Telefonía Fija', 'Wireless', 'Satelital']

# Se reemplaza 'Si' por 1 y '--' por 0 en las columnas seleccionadas
DF_Loc_con_int[tecnologias] = DF_Loc_con_int[tecnologias].replace({'Si': 1, '--': 0})

# Se convierte las columnas a tipo de datos numérico
DF_Loc_con_int[tecnologias] = DF_Loc_con_int[tecnologias].apply(pd.to_numeric, errors='coerce')

DF_Loc_con_int.head(4)
```

```python
# Se busca si existen filas que no tengan ninguna tecnologia 
tecnologia = ['ADSL', 'Cablemódem', 'Dial Up', 'Fibra Óptica', '4G', '3G', 'Telefonía Fija', 'Wireless', 'Satelital']

# Filtrar filas donde todas las columnas de interés son iguales a 0
filas_con_todo_cero = DF_Loc_con_int[(DF_Loc_con_int[tecnologia] == 0).all(axis=1)]
filas_con_todo_cero
```
